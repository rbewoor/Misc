## Solution 1: own approach without referring to the net
#### Checking division by 3 and then 5 allows every possible number to be checked with at most 3 IF conditions
#### Joining results and then printing once as i/o is slowest process generally. Also, RAM usage for 100 results is insignificant.

#### Referred https://pynative.com/python-random-randrange/ for random number gerenration

```python
import random
```

```python
def fizzbuzz(_n):
    """
    Implement fizzbuzz and return appropriate output
    Input: natural number
    Returns: A string with value "Fizz" / "Buzz" / "FizzBuss" / "input number itself"
    """
    if _n % 3 == 0:
        if _n % 5 == 0:
            # divisible by 3 and 5
            return('FizzBuzz')
        # divisible only by 3
        return('Fizz')
    if _n % 5 == 0:
        # divisible only by 5
        return('Buzz')
    # divisible by neither
    return(f"{_n}")
```

```python
## Run for first 100 natural numbers
results = list(map( fizzbuzz, list(range(1,101)) ))
results = '\n'.join(results)
print(results)
```

```python
## Run for 100 random natural numbers in some range
## Changed output format to show the random input number and its result
inp_arr = random.sample(range(1, 1000), 100)
results = list(map( fizzbuzz, inp_arr ))
for rand_n, res in zip(inp_arr, results):
    print(f"{rand_n} , {res}")
```

```python
def fizzbuzz_2(_n):
    """
    Implement fizzbuzz and return appropriate output
    Input: natural number
    Returns: A string with value "Fizz" / "Buzz" / "FizzBuss" / "input number itself"
    """
    answer = ''
    if _n % 3 == 0:
        answer = 'Fizz'
    if _n % 5 == 0:
        # divisible by 3 and 5
        answer = answer + 'Buzz'
        return(answer)
    ## new python feature that I saw in the youtube video but not used due to readability of code
    ## return(answer or f"{_n}")
    if answer == '':
        # divisible by neither
        answer = f"{_n}"
    return answer
```

```python
## Run for first 100 natural numbers
results_2 = list(map( fizzbuzz_2, list(range(1,101)) ))
results_2 = '\n'.join(results_2)
print(results_2)
```

## EDA

```python
def do_EDA(_indf, _col_list=None, _too_many_limit=40):
    """
    Do basic EDA: Info, Count of unique values, descriptive stats
    Input: 1) input dataframe, 2) list of columns to do EDA on (optional), 3) limit printing unique values
    Returns: Nothing
    """
    print(f"Total rows = {len(_indf)}\n")
    if _col_list is None:
        cols_to_check = _indf.columns
    else:
        cols_to_check = _col_list
    ## info
    print(f"\nInfo:\n{_indf.info()}")
    #print(f"")
    ## count unique and print values
    print(f"\nUnique values:")
    for idx, colname in enumerate(cols_to_check):
        cnt_uniq_vals = len(_indf[colname].unique())
        print(f"{idx+1:02}) {colname:12} : {cnt_uniq_vals:03} unique values - {_indf[colname].unique() if cnt_uniq_vals < _too_many_limit else 'Too many values'}")
    ## descriptive stats
    print(f"\nDescriptive stats:\n{_indf.describe()}")
```

#### Data preparation: using 1-hot coding for all factor variables
##### Reference Link: https://towardsdatascience.com/categorical-encoding-using-label-encoding-and-one-hot-encoder-911ef77fb5bd 

```python
def do_1Hotting(_in_df, _col_f_name, _prefix=None):
    """
    Create new columns for 1-hot encodings of the specified factor column, and reinserting the original column.
    Using pandas dummies values approach: i.e. pd.get_dummies
    Input: 1) dataframe 2) name of factor column to be 1-hot encoded 3) prefix for new column names
    Returns: dataframe with new columns added
    """
    ## original column is deleted by function, so first putting into temporary df and then readding the original column
    if _prefix is None:
        _prefix = _col_f_name
    temp_df = pd.get_dummies(_in_df, columns=[_col_f_name], prefix=[_prefix] )
    temp_df[_col_f_name] = _in_df[_col_f_name]
    return temp_df
```

#### Data preparation: Label Encoding for all factor variables
##### Reference Link: https://towardsdatascience.com/categorical-encoding-using-label-encoding-and-one-hot-encoder-911ef77fb5bd 

```python
def do_label_encoding(_in_df, _col_f_name, _prefix=None):
    """
    Create one new column for lanel encodings of the specified factor column, while retaining the original column values as is.
    Using sklearn.preprocessing.LabelEncoder()
    Input: 1) dataframe 2) name of factor column to be 1-hot encoded 3) prefix for new column names
    Returns: 1) dataframe with new column added 2) Mapping dictionary 3) label encoder object
    """
    sk_le = LabelEncoder()
    if _prefix is None:
        _prefix = _col_f_name
    new_col_name = _prefix + '_le'
    _in_df[new_col_name] = sk_le.fit_transform(_in_df[_col_f_name])
    
    le_name_mapping_dict = dict(zip(sk_le.classes_, sk_le.transform(sk_le.classes_)))
    return _in_df, le_name_mapping_dict, sk_le
```

### Creating new column called "G3_class" for G3 grade with three categories as follows:
#### G3 < 10       => Low
#### 10 <= G3 < 16 => Med
#### 16 <= G3      => High

```python
def create_G3_grade_categories(_in_df):
    """
    Will create a new column and set the value as a categorical value as follows:
    #### G3 < 10       => low
    #### 10 <= G3 < 16 => med
    #### 16 <= G3      => high
    Input: dataframe
    Return: dataframe with new columns added
    """
    _in_df['G3_class'] = 'Med'  ## Set everything to medium first
    _in_df.loc[_in_df['G3'] >= 16, 'G3_class'] = 'High'
    _in_df.loc[_in_df['G3'] < 10, 'G3_class'] = 'Low'
    return _in_df
```

```python
def visualize_tree_export_graphviz(_in_dt_classififer, _feature_cols, _class_names, _out_fname=None):
    """
    Visualize decision tree classifier passed by returning a plottable object and saving to png file
    Input: 1) tree classifier 2) input features (Xs) 3) Output lables (ys) 4) filename without .png extension
    Returns: Plottable graph object
    """
    dot_data = tree.export_graphviz(_in_dt_classififer, out_file=None,
                               feature_names = _feature_cols, class_names=_class_names,
                               filled=True, rounded=True, special_characters=True, )
    graph = graphviz.Source(dot_data, format="png")
    if _out_fname is not None:
        outfile_png = os.path.join(OUTDIR, _out_fname)
        graph.render(outfile_png)
        print(f"Created png file: {outfile_png}")
    return(graph)
```
